```python
import pandas as pd
from bs4 import BeautifulSoup
import urllib.request as urllib2
```

```python
#Region = []
Title = []
Date = []
#Author = []
Content = []

for i in range(1,11):
    url = "https://dominicanewsonline.com/news/?paged="+str(i)+"&s=hiv"
    html_page = urllib2.urlopen(url)
    #html_page = urllib2.urlopen(url)
    soup = BeautifulSoup(html_page)

# Extract the title of the article
    items = soup.find_all("article")
    #print(items)
    

    for item in items:
        #Reg = item.find("a", attrs={"class": "trb_search_result_section"})
        Tit = item.find("h2", attrs={"class": "entry-title"})
        Dat = item.find("div", attrs={"class": "entry-meta"})
        #Aut = item.find("", attrs={"class": "author"})
        Con = item.find("div", attrs={"class": "entry-content"})
        Date.append(Dat.text)
        #print(Dat)
        Title.append(Tit.text)
        #if Dat.has_attr('datetime'):
            #Date.append(Dat['datetime'])

        '''if Aut != None:
            Author.append(Aut.text)
        else:
            Author.append("None")'''
        if Con != None:
             Content.append(Con.text)
        else:
            Content.append("None")
    #print(len(Title))
    
```

```python
Content
```

```python
d = {'Title': Title, 'Date': Date, 'Content': Content}
df = pd.DataFrame(data=d)
```

```python
newDate = []
```

```python
for i in df['Date']:
    newDate.append(i[:i.find("|")])
    
```

```python
df['Date'] = newDate
```
