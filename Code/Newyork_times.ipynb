```python
from selenium import webdriver
from selenium.webdriver.common.keys import Keys
from bs4 import BeautifulSoup
import pandas as pd
```

```python
url = "https://www.nytimes.com/search?query=hiv&sort=best"
# Create Chrome Driver Session
driver = webdriver.Chrome(executable_path="C:\\Users\\User\\Downloads\\chromedriver_win32\\chromedriver.exe")
driver.implicitly_wait(10)
driver.get(url)

class_name = driver.find_element_by_xpath("//button[contains(.,'Show More')]") #Button Name
#Number of Page to see(Dynamic Page)
for _ in range(10):
    class_name.click()
    time.sleep(10)
```

```python
soup=BeautifulSoup(driver.page_source, 'lxml')

items = soup.find_all("div", attrs={"class": "css-138we14"})
Region = []
Title = []
Date = []
Author = []
Content = []

for item in items:
    Reg = item.find("p", attrs={"class": "css-myxawk"})
    Tit = item.find("h4", attrs={"class": "css-1lppelv"})
    Dat = item.parent.time.text
    Aut = item.find("p", attrs={"class": "css-15w69y9"})
    Con = item.find("p", attrs={"class": "css-1dwgixl"})
    Region.append(Reg.text)
    Title.append(Tit.text)
    Date.append(Dat)
    if Aut != None:
        Author.append(Aut.text)
    else:
        Author.append("None")
    Content.append(Con.text)
    
    
```

```python
d = {'Title': Title, 'Type/Region': Region, 'Date': Date, 'Author': Author, 'Content': Content}
df = pd.DataFrame(data=d)
```

```python
df.to_csv('Newyork_Times.csv')
```
