```python
import pandas as pd
from bs4 import BeautifulSoup
import urllib.request as urllib2
```

```python
#Region = []
Title = []
Date = []
Author = []
#Content = []

for i in range(1,11):
    url = "http://www.thestkittsnevisobserver.com/page/"+str(i)+"/?s=hiv"
    html_page = urllib2.urlopen(url)
    #html_page = urllib2.urlopen(url)
    soup = BeautifulSoup(html_page)

# Extract the title of the article
    items = soup.find_all("div", attrs={"class": "td-block-span6"})
    #print(items)
    

    for item in items:
        #Reg = item.find("a", attrs={"class": "trb_search_result_section"})
        Tit = item.find("h3", attrs={"class": "entry-title"})
        #print(Tit)
        Dat = item.find("span", attrs={"class": "td-post-date"})
        Aut = item.find("span", attrs={"class": "td-post-author-name"})
        #Con = item.find("div", attrs={"class": "jeg_post_excerpt"})
        Date.append(Dat.text)
        #print(Dat)
        Title.append(Tit.text)
        #print(Tit.text)
        #if Dat.has_attr('datetime'):
            #Date.append(Dat['datetime'])

        if Aut != None:
            Author.append(Aut.text)
        else:
            Author.append("None")
        '''if Con != None:
             Content.append(Con.text)
        else:
            Content.append("None")'''
    #print(len(Title))
   
```

```python
d = {'Title': Title, 'Date': Date, 'Author': Author}
df = pd.DataFrame(data=d)
```
