```python
import pandas as pd
from bs4 import BeautifulSoup
import urllib.request as urllib2
```

```python
#Region = []
Title = []
Date = []
#Author = []
Content = []

for i in range(1,11):
    url = "https://www.nowgrenada.com/page/"+str(i)+"/?s=hiv"
    html_page = urllib2.urlopen(url)
    #html_page = urllib2.urlopen(url)
    soup = BeautifulSoup(html_page)

# Extract the title of the article
    items = soup.find_all("article")
    #print(items)
    

    for item in items:
        #Reg = item.find("a", attrs={"class": "trb_search_result_section"})
        Tit = item.find("h3", attrs={"class": "jeg_post_title"})
        #print(Tit)
        Dat = item.find("div", attrs={"class": "jeg_meta_date"})
        #Aut = item.find("", attrs={"class": "author"})
        Con = item.find("div", attrs={"class": "jeg_post_excerpt"})
        Date.append(Dat.text)
        #print(Dat)
        Title.append(Tit.text)
        #print(Tit.text)
        #if Dat.has_attr('datetime'):
            #Date.append(Dat['datetime'])

        '''if Aut != None:
            Author.append(Aut.text)
        else:
            Author.append("None")'''
        if Con != None:
             Content.append(Con.text)
        else:
            Content.append("None")
    #print(len(Title))
   
```

```python
d = {'Title': Title, 'Date': Date, 'Content': Content}
df = pd.DataFrame(data=d)
```

```python
newTitle = []
newContent = []
```

```python
for i in df['Title']:
    i = i.replace('\n', '')
    i = i.replace('\t', '')
    newTitle.append(i)
```

```python
for i in df['Content']:
    i = i.replace('\n', '')
    i = i.replace('\t', '')
    newContent.append(i)
```

```python
d = {'Title': newTitle, 'Date': Date, 'Content': newContent}
df = pd.DataFrame(data=d)
```

```python
df.to_csv('Now_Grenada.csv',index=False)
```
