```python
import pandas as pd
from bs4 import BeautifulSoup
import urllib.request as urllib2
```

```python
Type = []
Title = []
Date = []
Author = []
#Content = []

for i in range(1,11):
    url = "https://www.theglobeandmail.com/search/?q=hiv%20aids&mode=news&page="+str(i)
    html_page = urllib2.urlopen(url)
    #html_page = urllib2.urlopen(url)
    soup = BeautifulSoup(html_page)

# Extract the title of the article
    items = soup.find_all("div", attrs={"class": "o-card--story"})
    #print(items)
    

    for item in items:
        Reg = item.find("span", attrs={"class": "c-label--card__content"})
        Tit = item.find("span", attrs={"class": "o-card__content-group"})
        #print(Tit)
        Dat = item.find("span", attrs={"class": "o-card__meta-time"})
        Aut = item.find("span", attrs={"class": "o-card__meta-authors"})
        #Con = item.find("div", attrs={"class": "td-excerpt"})
        Date.append(Dat.text)
        if Reg != None:
            Type.append(Reg.text)
        else:
            Type.append("None")
        
        #print(Dat)
        Title.append(Tit.text)
        #print(Tit.text)
        #if Dat.has_attr('datetime'):
            #Date.append(Dat['datetime'])

        if Aut != None:
            Author.append(Aut.text)
        else:
            Author.append("None")
        '''if Con != None:
             Content.append(Con.text)
        else:
            Content.append("None")
    #print(len(Title))'''
   
```

```python
d = {'Title': Title, 'Date': Date, 'Type': Type, 'Author': Author}
df = pd.DataFrame(data=d)
```

```python
df
```

```python
newTitle = []
newContent = []
```

```python
for i in df['Title']:
    i = i.replace('\n', '')
    i = i.replace('\t', '')
    newTitle.append(i)
```

```python
for i in df['Content']:
    i = i.replace('\n', '')
    i = i.replace('\t', '')
    newContent.append(i)
```

```python
d = {'Title': newTitle, 'Date': Date, 'Content': newContent}
df = pd.DataFrame(data=d)
```

```python
df.to_csv('The_globe_and_mail.csv',index=False)
```
