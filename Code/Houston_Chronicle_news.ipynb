```python
import pandas as pd
from bs4 import BeautifulSoup
import urllib.request as urllib2

# Opens the required website

Title = []
Date = []
Author = []
Content = []

for i in range(1,51):
    url = "https://www.chron.com/search/?action=search&searchindex=solr&query=hiv&isRedesign=&page="+str(i)
    html_page = urllib2.urlopen(url)
    soup = BeautifulSoup(html_page)

# Extract the title of the article
    items_has_photo = soup.find_all("div", attrs={"class": "gsa-item gsa-item-regular has-photo"})
    items_no_photo = soup.find_all("div", attrs={"class": "gsa-item gsa-item-regular no-photo"})
    

    for item in items_has_photo:
        Tit = item.find("h2", attrs={"class": "headline"})
        Dat = item.find('span',attrs={"class": "timestamp"})
        Aut = item.find("span", attrs={"class": "byline"})
        Con = item.find("p", attrs={"class": "blurb"})
        Title.append(Tit.text)
        if Dat != None:
            Date.append(Dat.text)
        else:
            Date.append("None")
        if Aut != None:
            Author.append(Aut.text)
        else:
            Author.append("None")
        if Con != None:
             Content.append(Con.text)
        else:
            Content.append("None")
    
        for item in items_no_photo:
            Tit = item.find("h2", attrs={"class": "headline"})
            Dat = item.find('span',attrs={"class": "timestamp"})
            Aut = item.find("span", attrs={"class": "byline"})
            Con = item.find("p", attrs={"class": "blurb"})
            Title.append(Tit.text)
            if Dat != None:
                Date.append(Dat.text)
            else:
                Date.append("None")
            if Aut != None:
                Author.append(Aut.text)
            else:
                Author.append("None")
            if Con != None:
                Content.append(Con.text)
            else:
                Content.append("None")
            
        

        
    
```

```python
d = {'Title': Title, 'Date': Date, 'Author': Author, 'Content': Content}
df = pd.DataFrame(data=d)
```

```python
df
```

```python
newTitle = []
newContent = []
```

```python
for i in df['Title']:
    i = i.replace('\n', '')
    i = i.replace('\t', '')
    newTitle.append(i)
```

```python
for i in df['Content']:
    i = i.replace('\n', '')
    i = i.replace('\t', '')
    newContent.append(i)
```

```python
d = {'Title': newTitle, 'Date': Date,  'Author': Author, 'Content': newContent}
df = pd.DataFrame(data=d)
```

```python
df = df.drop_duplicates(subset=None, keep='first', inplace=False)
```
