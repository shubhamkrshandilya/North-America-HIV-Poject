```python
import pandas as pd
from bs4 import BeautifulSoup
import urllib.request as urllib2
```

```python
Type = []
Title = []
Date = []
#Author = []
Content = []
```

```python
i = 0
url1 = "https://www.trinidadexpress.com/search/?f=html&q=hiv&sd=desc&l=100&t=article%2Ccollection&nsa=eedition"
url2 = "https://www.trinidadexpress.com/search/?f=html&q=hiv&sd=desc&l=100&t=article%2Ccollection&nsa=eedition&app%5B0%5D=editorial&o=100"
urls = [url1,url2]
for url in urls:
    
    html_page = urllib2.urlopen(url)
    #html_page = urllib2.urlopen(url)
    soup = BeautifulSoup(html_page)

    # Extract the title of the article
    items = soup.find_all("article")
    #print(items)


    for item in items:
        print(i)
        i =+ 1
        Reg = item.find("div", attrs={"class": "card-labels"})
        Tit = item.find("div", attrs={"class": "card-headline"})
        #print(Tit)
        Dat = item.find("li", attrs={"class": "card-date"})
        #Aut = item.find("", attrs={"class": "author"})
        Con = item.find("div", attrs={"class": "card-lead"})
        if Tit == None:
            continue
        Title.append(Tit.text)
        if Dat != None:
            Date.append(Dat.text)
        else:
            Date.append("None")
        #print(Dat)
        
        
        #print(Tit.text)
        #if Dat.has_attr('datetime'):
            #Date.append(Dat['datetime'])

        if Reg != None:
            Type.append(Reg.text)
        else:
            Type.append("None")
        if Con != None:
             Content.append(Con.text)
        else:
            Content.append("None")
#print(len(Title))
```

```python
d = {'Title': Title, 'Date': Date, 'Type': Type, 'Content': Content}
df = pd.DataFrame(data=d)
```

```python
newTitle = []
newContent = []
newDate  = []
newType = []
```

```python
for i in df['Title']:
    i = i.replace('\n', '')
    i = i.replace('\t', '')
    newTitle.append(i)
```

```python
for i in df['Content']:
    i = i.replace('\n', '')
    i = i.replace('\t', '')
    newContent.append(i)
```

```python
for i in df['Type']:
    i = i.replace('\n', '')
    i = i.replace('\t', '')
    newType.append(i)
```

```python
for i in df['Date']:
    i = i.replace('\n', '')
    i = i.replace('\t', '')
    newDate.append(i)
```

```python
d = {'Title': newTitle, 'Date': newDate, 'Type': newType, 'Content': newContent}
df = pd.DataFrame(data=d)
```

```python
df.to_csv('Trinidad_express.csv',index=False)
```
