```python
import pandas as pd
from bs4 import BeautifulSoup
import urllib.request as urllib2
```

```python
#Region = []
Title = []
Date = []
Author = []
Content = []

for i in range(1,21):
    url = "http://www.nationnews.com/search/hiv?facet=true&page="+str(i)+"&sort=Relevance&utf8=%E2%9C%93"
    html_page = urllib2.urlopen(url)
    soup = BeautifulSoup(html_page)

# Extract the title of the article
    items = soup.find_all("div", attrs={"class": "list-content"})
    #print(items)
    

    for item in items:
        #Reg = item.find("a", attrs={"class": "trb_search_result_section"})
        Tit = item.find("h3")
        Dat = item.find("div", attrs={"class": "publish-date"})
        Aut = item.find("div", attrs={"class": "author"})
        Con = item.find("div", attrs={"class": "list-bottom-text"})
        Date.append(Dat.text)
        #print(Dat)
        Title.append(Tit.text)
        #if Dat.has_attr('datetime'):
            #Date.append(Dat['datetime'])

        if Aut != None:
            Author.append(Aut.text)
        else:
            Author.append("None")
        if Con != None:
             Content.append(Con.text)
        else:
            Content.append("None")
    #print(len(Title))
    
```

```python
d = {'Title': Title, 'Date': Date, 'Author': Author, 'Content': Content}
df = pd.DataFrame(data=d)
```

```python
newTitle = []
newDate = []
newAuthor = []
newContent = []
```

```python
for i in df['Title']:
    i = i.replace('\n', '')
    i = i.replace('\t', '')
    newTitle.append(i)
```

```python
for i in df['Date']:
    i = i.replace('\n', '')
    i  = i.replace('\t', '')
    i  = i.replace('Added', '')
    newDate.append(i)
```

```python
for i in df['Author']:
    i = i.replace('\n', '')
    i = i.replace('\t', '')
    i = i.replace('By', '')
    newAuthor.append(i)
```

```python
for i in df['Content']:
    i = i.replace('\n', '')
    i = i.replace('\t', '')
    newContent.append(i)
```

```python
d = {'Title': newTitle, 'Date': newDate, 'Author': newAuthor, 'Content': newContent}
df = pd.DataFrame(data=d)
```

```python
df.to_csv('Nation_News_Barbados.csv',index=False)
```
