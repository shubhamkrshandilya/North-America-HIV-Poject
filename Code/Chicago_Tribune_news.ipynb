```python
import pandas as pd
from bs4 import BeautifulSoup
import urllib.request as urllib2

# Opens the required website
Region = []
Title = []
Date = []
Author = []
Content = []

for i in range(1,11):
    url = "https://www.chicagotribune.com/search/dispatcher.front?page="+str(i)+"&target=all&spell=on&Query=hiv#trb_search"
    html_page = urllib2.urlopen(url)
    soup = BeautifulSoup(html_page)

# Extract the title of the article
    items = soup.find_all("div", attrs={"class": "trb_search_result"})
    

    for item in items:
        Reg = item.find("a", attrs={"class": "trb_search_result_section"})
        Tit = item.find("a", attrs={"class": "trb_search_result_title"})
        Dat = item.find('time')
        Aut = item.find("span", attrs={"class": "trb_search_result_author_nm_pm"})
        Con = item.find("div", attrs={"class": "trb_search_result_description"})
        Region.append(Reg.text)
        Title.append(Tit.text)
        if Dat.has_attr('datetime'):
            Date.append(Dat['datetime'])

        if Aut != None:
            Author.append(Aut.text)
        else:
            Author.append("None")
        if Con != None:
             Content.append(Con.text)
        else:
            Content.append("None")
        
    
```

```python
d = {'Title': Title, 'Type/Region': Region, 'Date': Date, 'Author': Author, 'Content': Content}
df = pd.DataFrame(data=d)
```

```python
df.to_csv('Chicago_Tribune.csv')
```
